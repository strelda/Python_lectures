### Find all divisors
Find all divisors of a number using previously written heart of the function.

```python 
# input number
n = 33

# find the smallest divisor of n
d = 2
while d < n:
    if n%d == 0:
        print(n, "can be divided by", d)
        break
    d += 1
else:
    print(n, "is prime")
```
