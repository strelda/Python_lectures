````python
from enum import Enum

class Light(Enum):
    RED = 1
    ORANGE = 2
    GREEN = 3

def next_light(current_light:Light)->Light:
    if current_light == Light.RED:
        return Light.ORANGE
    elif current_light == Light.ORANGE:
        return Light.GREEN
    else:
        return Light.RED
 
current = Light.RED # same as TrafficLight(1)
print(next_light(current))
````
