# Tuples

### Motivation
Tuples are used when we want to group elements together, but we do not want to change them. This can be useful when we want to ensure that the data is not changed by accident, or when Python requires an immutable type (for example, as keys in a dictionary).

```python
coordinate =  [4,5,1]
coordinate[1] = 3
print(coordinate)
```

```python
immortal_coordinate =  (4,5,1)
immortal_coordinate[1] = 3
```

```python
# in a snake game, we can have obstacles and a list of messages that appear at each of them
obstacles = {
    (0,0): "you are at the origin",
    (3,1): "You like stones",
    (4,1): "You like stones"
}
obstacles[(0,0)]
```

```python
obstacles = {
    [0,0]: "blah"
}
obstacles[[0,0]]
```

Ans here we are at the `hashing` again! Python needs to be able to search for the `key` in the `dictionary` quickly, thus it needs to hash it. Mutable types can change, thus their hash can change, thus they cannot be used as keys in dictionaries.

## Introduction

```python
t = (1,2,3) # equivalent to t = 1,2,3
print(t)
```

This can be used for saving more variables at once, switching them...

```python
a,b = 1,2
a,b = b,a # swap values
print(a)
print(b)
```

```python
print(t)
a,b,c = t # unpacking tuple t to a,b,c
print(a)
```

Also can be used inside functions for returning more than one value:

```python
def complex(real: float, imag: float) -> tuple[float,float]:
    return real, imag

print(complex(1.,2.))
```

Conversions between lists and tuples are done using `list()` and `tuple()`:

```python
print(t)
print(list(t))
print(tuple(list(t)))
```

## + Unpacking 
We can even put more variables into one cell using `*args` (arguments). Here we can pass as many arguments as we want to the function. They are accepted as tuple.

```python
def f_args(*a):
    return a[0]
print(f_args(1,2,3))
```

Using double star as `**kwargs` (keyword arguments), we pass the arguments as a dictionary.

```python
def f_kwargs(**kwargs):
    print(kwargs)

f_kwargs(a=1,b=2,c=3)
f_kwargs(**{'a':1,'b':2,'c':3})
```

Unpacking can be used even for lists, as shown below. This can be handy for example for going through a list using recursion. Remember the sorting or searching algorithms we did?

```python
lst = [1, 2, 3, 4, 5]
first, *rest = lst
print(first)
print(rest)
```
