```python
class NodeBinary:
    """A node in a binary tree. Each node has a value and references to its left and right children."""
    def __init__(self, x):
        self.info = x      # value of the node
        self.left = None   # left child
        self.right = None  # right child
    def preorder(self):
        """Print the value of the node and then recursively print the values of the left and right subtrees."""
        print(self.info)
        if self.left:
            self.left.preorder()
        if self.right:
            self.right.preorder()
    def inorder(self):
        """Recursively print the values of the left subtree, then the value of the node, then the values of the right subtree."""
        if self.left:
            self.left.inorder()
        print(self.info)
        if self.right:
            self.right.inorder()
    def postorder(self):
        """Recursively print the values of the left and right subtrees, then the value of the node."""
        if self.left:
            self.left.postorder()
        if self.right:
            self.right.postorder()
        print(self.info)
```

```python
tree = NodeBinary("I'm Root!")
tree.left = NodeBinary("left child")
tree.right = NodeBinary("right child")

print("Preorder:")
print(tree.preorder())

print("\nInorder:")
print(tree.inorder())

print("\nPostorder:")
print(tree.postorder())
```

Notice that the difference is only in the position of the `print` statement. It is before/between/after we recursively call for subtrees.

### General trees
Now the tree has any number of children, lets save them into a list

```python
class Node:
    """A node of a general tree."""
    def __init__(self, x):
        self.info = x
        self.children = []
    def preorder(self):
        """Print the value of the node and then recursively print the values of the subtrees."""
        print(self.info)
        for child in self.children:
            child.preorder()
```

```python
root = Node("I'm Root!")
root.children.append(Node(2))
root.children.append(Node(3))

child1 = root.children[0]
child1.children.append(Node(4))
child1.children.append(Node(5))

print("Preorder:")
print(root.preorder())
```
