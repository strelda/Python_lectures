# Graphs

- **node**: a point in the graph
- **edge**: a connection between two nodes
- **directed graph**: a graph where the edges have a direction
- **weighted graph**: a graph where the edges have weights
- **adjacency matrix**: a matrix with the rows and columns representing the nodes and the values representing the edges
- **degree**: the number of edges connected to a node
- **path**: a sequence of nodes where each node is connected to the next node
- **cycle**: a path that starts and ends at the same node
- **connected graph**: a graph where there is a path between every pair of nodes

```python
class Graph:
    """
    A simple undirected graph implementation using an adjacency list.
    The graph is represented as a dictionary where keys are node identifiers and values are lists of adjacent nodes.
    """
    def __init__(self):
        self.graph = {}

    def add_edge(self, u, v):
        # create a key if it doesn't exist
        if u not in self.graph:
            self.graph[u] = []
        if v not in self.graph:
            self.graph[v] = []
        # append connections to the lists
        self.graph[u].append(v)
        self.graph[v].append(u)

    def remove_edge(self, u, v):
        if u in self.graph and v in self.graph[u]:
            self.graph[u].remove(v)
        if v in self.graph and u in self.graph[v]:
            self.graph[v].remove(u)

    def adjacency_matrix(self):
        nodes = list(self.graph.keys())
        size = len(nodes)
        matrix = [[0] * size for _ in range(size)]
        for i in range(size):
            for j in range(size):
                if nodes[j] in self.graph[nodes[i]]:
                    matrix[i][j] = 1
        return matrix

    def __repr__(self):
        return str(self.graph)

# Example usage
g = Graph()
g.add_edge(1, 2)
g.add_edge(2, 3)
g.add_edge(3, 1)
print(g)
g.adjacency_matrix()
```

## igraph example
For installation go into the terminal: `pip install igraph`

```python
import igraph as ig
g = ig.Graph(directed=False)
g.add_vertices(5)
g.add_edges([(0, 1), (1, 2), (2, 0), (0, 2), (4, 0), (1, 3)])
g.vs["label"] = ["A", "B", "C", "D", "E","F"]
print(g)
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ig.plot(g, target=ax)
plt.show()
```

```python
# Get the shortest path from D (vertex 3) to E (vertex 4)
shortest_path = g.get_shortest_paths(3, 4)[0]  # Need [0] because get_shortest_paths returns a list of paths
print(shortest_path)
def plot_g_with_path(g, path):
    path_edges = [(path[i], path[i+1]) for i in range(len(path)-1)]
    # Create a list of edge colors
    edge_colors = ['lightgray'] * g.ecount()
    # Set the color of the path edges to red
    for edge_id in [g.get_eid(source, target) for source, target in path_edges]:
        edge_colors[edge_id] = 'red'

    fig, ax = plt.subplots()
    ig.plot(g, target=ax, edge_color=edge_colors)
    plt.show()

plot_g_with_path(g, shortest_path)
```

### Directed graph

```python
g = ig.Graph(directed=True)
g.add_vertices(4)
g.add_edges([(0, 1), (1, 2), (2, 0), (0, 2), (1, 3)])

fig, ax = plt.subplots()
ig.plot(g, target=ax)
plt.show()
```

### Weighted graph

```python
g = ig.Graph(directed=True)
g.add_vertices(4)
g.add_edges([(0, 1), (1, 2), (2, 0), (0, 2), (1, 3)])
g.es[0]["weight"] = 1
g.es[1]["weight"] = 3
g.es[2]["weight"] = 6
g.es[3]["weight"] = 9

# plot with weights using thickness
fig, ax = plt.subplots()
ig.plot(g, target=ax, edge_width=g.es["weight"])
plt.show()
```
