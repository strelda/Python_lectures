# Implement some of them into the one-sided LinkedList

```python
def prev_node(self, x): 
    """Find the node with value x and return the node before it. Return None the node with x is the first."""
    ...
def succ_node(self, n):
    """Return the node after the one with index n. Return None if n is the last node."""
    ...
def reverse(self): 
    """Change the list so that the order of the nodes is reversed."""
    ...
```

```python
class MyNode:
    def __init__(self, val):
        # Prvek si pamatuje hodnotu a odkaz na předchůdce i následníka
        self.value = val
        self.prev = None
        self.next = None

class LinkedList:
    def __init__(self):
        # we hold the list using a head node, has value None
        self.head = MyNode(None)
        self.head.prev = self.head.next = self.head

    def insert_node(self, i, left):
        """Insert a node after index `n`."""
        right = left.next
        i.prev = left
        i.next = right
        right.prev = i
        left.next = i

    def prepend_node(self, i):
        """Prepend node i to the beginning - after the head"""
        self.insert_node(i, self.head)

    def append_node(self, i):
        """Append node i to the end before the head, i.e. after the element before the head."""
        self.insert_node(i, self.head.prev)

    def prepend(self, val):
        """Add a node with a value val to the beginning of the list."""
        return self.prepend_node(MyNode(val))

    def append(self, val):
        """Zapojí prvek se zadanou hodnotu na konec seznamu."""
        return self.append_node(MyNode(val))

    def print(self):
        n = self.head.next
        while n != self.head:
            print(n.value)
            n = n.next
            
    def find(self, x):
        """Return the node with value x. Return None if not found."""
        current = self.head.next
        while current != self.head:
            if current.value == x:
                return current
            current = current.next
        return None

    def remove_node(self, i):
        """Remove the node at index i."""
        if i < 0:
            return
        current = self.head.next
        count = 0
        while current != self.head:
            if count == i:
                current.prev.next = current.next
                current.next.prev = current.prev
                return
            count += 1
            current = current.next

    def prev_node(self, x):
        """Find the node with value x and return the node before it. Return None if the node with x is the first."""
        current = self.head.next
        while current != self.head:
            if current.value == x:
                if current.prev != self.head:
                    return current.prev
                else:
                    return None
            current = current.next

    def succ_node(self, n):
        """Return the node after the one with index n. Return None if n is the last node."""
        if n < 0:
            return None
        current = self.head.next
        count = 0
        while current != self.head:
            if count == n:
                if current.next != self.head:
                    return current.next
                else:
                    return None
            count += 1
            current = current.next

    def reverse(self):
        """Change the list so that the order of the nodes is reversed."""
        current = self.head
        while True:
            current.prev, current.next = current.next, current.prev
            current = current.prev  # Move to the next node in the reversed order
            if current == self.head:
                break
```
