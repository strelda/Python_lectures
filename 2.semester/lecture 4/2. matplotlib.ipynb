# Matplotlib

```python
import matplotlib.pyplot as plt
import numpy as np
```

# Lines

Matplotlib by default works with data in the form of lists `x = [1, 2, 3]`, or numpy arrays `x = np.array([1,2,3])`.

```python
x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('matplotlib.pyplot')
plt.show()
```

First `plt.plot()` creates a canvas and following plots are drawn on it, until `plt.show()` or `plt.savefig()` is called.

```python
x = np.linspace(0, 10, 30)
y1 = np.sin(x)
y2 = np.tan(x)
y3 = np.cos(x)

plt.plot(x, y1, 'bo-', label='sin(x)') # blue b, circle o, solid line -
plt.plot(x, y2, 'r--', label='tan(x)') # red r, dashed line --
plt.plot(x, y3, 'g^' , label='cos(x)') # green g, triangle up ^
plt.legend()
plt.ylim(-3,3)
plt.xlabel('x')
plt.ylabel('f(x)')
plt.savefig('sin_tan.pdf')
```

```python
x_l = np.logspace(1, 10, 100)
y_l = np.log(x_l)
plt.plot(x_l, y_l)
plt.loglog() # both axis in log
# plt.semilogx() # only x axis in log
# plt.semilogy() # only y axis in log
plt.yticks([1, 10, 100, 1000])
plt.xticks([10**i for i in range(11)])
plt.show()
```

# Scatter
https://matplotlib.org/stable/gallery/shapes_and_collections/scatter.html#scatter-plot
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html#matplotlib.pyplot.scatter


```python
x_p = np.linspace(0, 10, 50)
y_p = np.random.random(50)

plt.scatter(x_p, y_p, marker='x', c=[[0,0,i/10] for i in x_p])
plt.hlines(0.5, xmin=0, xmax=10, linestyles='dashed') # draw a horizontal line
plt.hlines([0.25,0.75], 0, 10, colors='green') # y coordinate can be added as a list, xmin and xmax can be used as positional arguments
plt.annotate('here', (6, 0.1), xytext=(7, 0.2), arrowprops=dict(facecolor='black', shrink=0.05), fontsize=20) # put a text with an arrow on a coordinate
plt.text(2, 0.1, 'and here', fontsize=20) # put only text on a coordinate
plt.show()
```

# Histogram
https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html#matplotlib.pyplot.hist

```python
x_1 = np.random.normal(0, 1, 1000)
x_2 = np.random.normal(0, 1, 1000)

plt.hist(x_1, bins=20, color='red', alpha=0.5, label='x')
plt.hist(x_2, bins=100, color='blue', alpha=0.5, label='y')
plt.title('random numbers histogram')
plt.xlabel('random variable')
plt.ylabel('frequency')
plt.legend()
plt.show()
```

# Subplots

```python

plt.subplot(2, 2, 1) # number of rows, number of columns, plot index
plt.plot(x, y1, 'b', label='sin(x)')
plt.ylabel("kaoda")
plt.xlabel("askfhaa")
plt.subplot(2, 2, 2)
plt.plot(x, y2, 'r', label='tan(x)')

plt.subplot(2, 2, 3)
plt.plot(x, y3, 'g' , label='cos(x)')

plt.subplot(2, 2, 4)
plt.plot(x, y1, 'b', label='sin(x)')
plt.plot(x, y2, 'p', label='sin(x)')

plt.show()
```
