# Numerical derivative
1. Newton's difference quotient
2. Central difference quotient
3. Scipy's derivative function

```python
f = lambda x: x**2  # Function: f(x) = x^2
x0 = 2
```

```python
def derivative(f, x, dx=1e-5):
    return (f(x + dx) - f(x)) / dx

print(derivative(f, x0))  # Approximate derivative at x=2
```

<div class="alert alert-block alert-warning">
Beware of small step size. Going to zero does not reduce the error in floating point arithmetics.
</div>

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-20, 0, 1000)
y1 = derivative(f, 2, dx=pow(10,x))
plt.plot(x, y1)
plt.xlabel('log10(h)')
plt.ylabel('Approximate derivative')
plt.show()
```

```python
import jax.numpy as jnp
from jax import grad

dfdx = grad(f)(2.0)

print(dfdx)  # Output: 4.0
```
