### GitHub Copilot in VS Code

GitHub Copilot is an AI coding assistant that suggests code as you type. It can:
- Complete lines and entire functions
- Generate code from comments
- Explain existing code
- Help with debugging

**How to use it effectively:**
1. Write clear comments describing what you want
2. Start typing and let Copilot suggest completions
3. Press `Tab` to accept, or `Esc` to dismiss
4. Use `Ctrl+Enter` to see multiple suggestions

## Best practices for AI-assisted coding
Learn to work with the specific tool you are using. They use different tips, such as `CLAUDE.md` files for Claude Code.
- be specific
- provide context
- always verify AI-generated code and tests

links: 
- [How Anthropic teams use Claude Code](https://claude.com/blog/how-anthropic-teams-use-claude-code)
- [How AI is transforming work at Anthropic](https://www.anthropic.com/research/how-ai-is-transforming-work-at-anthropic)

**Beware using it for security parts of the code. Usually safe for frontend, but not always for backend.**

### 1. Level: Friend on the phone
- Use GPT instead of browser or Documentation for speed.
- Debugging: ask GPT to suggest minor fixes or explain error messages.
- Ask for small advices and suggesting options on how to continue when you are stuck. Use socratic method: ask GPT to ask you questions about the problem, so you can find the solution yourself. Thiss way you will not stop thinking.
- Learn about someone else's code

### 2. Level: Coding assistant
- Improve your code, not the algorithm.
- Prompt AI to generate tests for your code. Don't forget to review the tests.

**These correspond to steps 5. and 6. in the How to design functions process.**

### 3. Level: Workflow for small tasks
- Think about the structure of the code. 
- Write down wishlist and first 2. steps of How to design functions process.
- Let GPT finish the rest:
    1. write tests,
    2. write functional examples,
    3. write the function itself.

*It might seem counterintuitive to write tests before the function itself, but turns out that it often works better.*

### 4. Level: Workflow for entire project
- Define clearly what you want from the project and provide all context (scientific articles, documentation, etc.)
- Use GPT to generate plan (if available in the coding assistant) and review it.
- If the task is hard, let it generate pseudocode and tests (again tests before the code).
- Review the pseudocode and tests, improve the algorithms or general structure.
- Use GPT to generate the actual code based on the pseudocode

*If the result is wrong, starting from the beginning again is often faster than trying to fix a messy code.*
