### Divide-and-conquer vs. dynamic programming
Divide-and-conquer method takes advantage of the problem reducibility. We divide the complex problem into many simple tasks and solve them independently. In recursion, we did something similar. We divided the problem into subproblems and solved them recursively.

Using dynamic programming, we can look at the problem, precalculate some smaller steps (save them to the memory), and at the end, put all together.
